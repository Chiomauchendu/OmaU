**Data Cleaning and Preprocessing**

```python
X = data.drop("HEATING_COST", axis=1)
y = data["HEATING_COST"]
```

```python
X = X.dropna()  # Remove rows with missing values
y = y[X.index]
```

```python
numerical_cols = X.select_dtypes(include=['float64']).columns
categorical_cols = X.select_dtypes(include=['object']).columns
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X[numerical_cols] = imputer.fit_transform(X[numerical_cols])
```

**Encoded the data**

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
encoder = OneHotEncoder(drop='first', sparse=False)
ct = ColumnTransformer(
    transformers=[('encoder', encoder, categorical_cols)],
    remainder='passthrough'
)
X_encoded = ct.fit_transform(X)
```

```python
print(X_encoded)
```

**Standard Scalar**

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)
```

**Splitting the data**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape,X_test.shape,X_scaled.shape)
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg_model=LogisticRegression()
```

```python
lg_model.fit(X_train,y_train)
```

```python
y_pred=lg_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
r2_value=r2_score(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
```

```python
print("The Logistic Regression r2 score is :",r2_value)
print("The logistic Regression MAE is:",mae)
print("The logistic Regression MSE is:",mse)
```
